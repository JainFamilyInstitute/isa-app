```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
# rename columns
df.rename(columns={'Quantiles':'quantile'}, inplace=True)
df.rename(columns={'Variables':'variable'}, inplace=True)
df.rename(columns={'Borrowing Amount':'amount'}, inplace=True)
df.rename(columns={'Financial Instrument':'type'}, inplace=True)
df.rename(columns={'Risk Aversion':'risk'}, inplace=True)
```

```python
df.head()
```

```python
# recode quantile values
df.loc[df['quantile'] == "1st quantile", 'quantile'] = 1
df.loc[df['quantile'] == "2nd quantile", 'quantile'] = 2
df.loc[df['quantile'] == "3rd quantile", 'quantile'] = 3
df.loc[df['quantile'] == "4th quantile", 'quantile'] = 4
df.loc[df['quantile'] == "5th quantile", 'quantile'] = 5
```

```python
df.head()
```

```python
# get columns with dollar amounts to round
cols = []
for x in range(22, 101):
    value = x
    cols.append(str(x))
print(cols)
```

```python
# round selected columns' values to nearest whole number
df[cols] = df[cols].round()
```

```python
# see what values there are to filter rows by
quantiles = df['quantile'].unique()
variables = df['variable'].unique()
amounts = df['amount'].unique()
types = df['type'].unique()
risks = df['risk'].unique()
print("Quantiles: " + str(quantiles))
print("\t")
print("Variables: " + str(variables))
print("\t")
print("Amount: " + str(amounts))
print("\t")
print("Type: " + str(types))
print("\t")
print("Risk: " + str(risks))
```
